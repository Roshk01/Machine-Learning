```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_digits
```

```python
datadigit = load_digits()
dir(datadigit)
```

```python
plt.matshow(datadigit.images[24]);
```

```python
datadigit.target[24]
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(datadigit.data, datadigit.target, train_size = 0.9)
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(x_train,y_train)
```

```python
# lr.predict(x_test)
plt.matshow(datadigit.images[89])
```

```python
datadigit.target[9]
```

```python
lr.predict([datadigit.data[9]])
```

```python
lr.score(x_test,y_test)
```

```python
y_predict = lr.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test,y_predict)
```

```python
import seaborn as sns
plt.figure(figsize =(10,6))
sns.heatmap(cm, annot = True)
plt.xlabel('predicted_value')
plt.ylabel('actual_value')
```
