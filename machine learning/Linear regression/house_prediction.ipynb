```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
data = pd.read_csv("homepricesonefeature.csv")
data.head()
```

```python
%matplotlib inline
plt.title("House Pricing")
plt.xlabel(" Area")
plt.ylabel("Price")
plt.scatter(data.area,data.price, color = "red", linewidth = 4)
```

```python
reg = LinearRegression()
reg.fit(data[['area']],data.price)
```

```python
reg.predict([[2300]])
```

```python
print(" coeficient of price: ", reg.coef_)
print(" intercept of price: ", reg.intercept_)
```

# linear regression of prediction is:
    Linear Regression Formula: 
        Y = mx + b  ( slope Equation )
        Price = m * area + b
        
    m = coeficient of price 
    b = intercept of price 

```python
m = 135.78767123
b = 180616.43835616432
area = 2300
m * area + b
```

# Both the values through formula and linear regression function are same 

```python
%matplotlib inline
plt.title("house Prediction")
plt.xlabel("Area")
plt.ylabel("price")
plt.scatter(data.area,data.price, color= 'red')
plt.plot(data.area,reg.predict(data[['area']]))
```
