```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("income.csv")
len(data)
```

```python
x = data.iloc[:, [1, 2]].values  
y = data['Income($)']
```

```python
from sklearn.cluster import KMeans
wcss = []
for i in range(1,11):
    kmeans = KMeans(n_clusters = i, init='k-means++', random_state=42, n_init=10)
    kmeans.fit(x)
    wcss.append(kmeans.inertia_)
    
```

```python
plt.plot(range(1,11), wcss)
plt.title('finding clustor')
plt.xlabel('Range')
plt.ylabel('wcss list')
plt.show()
```

```python
plt.scatter(data['Age'],data['Income($)'])
plt.title('visual representation of data')
plt.xlabel('Age')
plt.ylabel('Income($)')
plt.show()
```

```python
kmeans = KMeans(n_clusters = 3 ,init='k-means++', random_state=42, n_init=10)
kmeans
```

```python
y_predict = kmeans.fit_predict(data[['Age','Income($)']])
y_predict
```

```python
data['clusters'] = y_predict
data.head()
```

```python
c1 = data[data.clusters==0]
c2 = data[data.clusters==1]
c3 = data[data.clusters==2]
plt.scatter(c1['Age'],c1['Income($)'], color = 'green', label = 'Income($)')
plt.scatter(c2['Age'],c2['Income($)'], color = 'red', label = 'Income($)')
plt.scatter(c3['Age'],c3['Income($)'], color = 'yellow', label = 'Income($)')
plt.xlabel('Age')
plt.ylabel('Income($)')
plt.legend()
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaler.fit(data[['Income($)']])
data['Income($)']= scaler.transform(data[['Income($)']])

scaler.fit(data[['Age']])
data['Age']= scaler.transform(data[['Age']])
data.head()
```

```python
kmeans = KMeans(n_clusters = 3 ,init='k-means++', random_state=42, n_init=10)
y_predict = kmeans.fit_predict(data[['Age','Income($)']])
y_predict
```

```python
data['clusters'] = y_predict
data.head()
```

```python
kmeans.cluster_centers_
```

```python
c1 = data[data.clusters==0]
c2 = data[data.clusters==1]
c3 = data[data.clusters==2]
plt.scatter(c1['Age'],c1['Income($)'], color = 'green', label = 'Income($)')
plt.scatter(c2['Age'],c2['Income($)'], color = 'red', label = 'Income($)')
plt.scatter(c3['Age'],c3['Income($)'], color = 'yellow', label = 'Income($)')
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1], color = 'blue', marker = '*')
plt.xlabel('Age')
plt.ylabel('Income($)')
plt.legend()
```
